```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mplc
import matplotlib
matplotlib.rc('font', family='Arial')

# matrix = matrix[:9,:9]
```

```python
# Notes, the double si, sì sí is great as is the yóò context!
matrix = np.load("./toprint/emi_yoo_si_si_oju_mi_si_ile_juda_emi_yoo_si_bu_ifoju_lu_gbogbo_esin_ti_orile-ede.npy")
print(np.shape(matrix))
matrix = matrix[:9,:9]
plt.figure(figsize=(10,10))
# plt.imshow(matrix, cmap='afmhot', interpolation="None", norm=mplc.LogNorm(vmin=0.001, vmax=matrix.max()))
plt.imshow(matrix, cmap='gray', interpolation="None", norm=mplc.LogNorm(vmin=0.001, vmax=matrix.max()))

# plt.colorbar()

# my_xticks = "emi yoo si si oju mi si ile juda emi yoo si bu ifoju lu gbogbo esin ti orile-ede".split(" ")
# my_yticks = "èmi yóò sì sí ojú mi sí ilé júdà èmi yóò sì bu ìfọ́jú lu gbogbo ẹsin ti orílẹ̀-èdè".split(" ")

my_xticks = "emi yoo si si oju mi si ile juda".split(" ")
my_yticks = "èmi yóò sì sí ojú mi sí ilé júdà".split(" ")

ax = plt.gca()
ax.set_xticks(np.arange(0, len(my_xticks), 1))
ax.set_yticks(np.arange(0, len(my_yticks), 1))
ax.xaxis.tick_top()

plt.tick_params(axis='both', which='major', labelsize=25)
plt.tick_params(axis='both', which='minor', labelsize=1)

plt.xticks(np.arange(0, len(my_xticks), 1), my_xticks)
plt.yticks(np.arange(0, len(my_yticks), 1), my_yticks)
plt.show()
```
